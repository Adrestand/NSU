```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
wine = fetch_ucirepo(id=109)

# data (as pandas dataframes)
X = wine.data.features
y = wine.data.targets
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
# создание DataFrame из признаков и целевой переменной
wine_df = pd.DataFrame(data=X, columns=wine.feature_names)
wine_df['target'] = y

# отображение первых 5 строк DataFrame wine
print("Первые 5 строк DataFrame wine:")
print(wine_df.head())

# отображение последних 5 строк DataFrame wine
print("\nПоследние 5 строк DataFrame wine:")
print(wine_df.tail())
```

```python
wine_df.to_excel('wine_data.xlsx', index=False, float_format='%.2f', engine='xlsxwriter')
```

```python
from scipy.stats import kstest

# Проход по всем столбцам данных
for column in wine_df.columns:
    # Проверка на нормальность распределения с помощью критерия Колмогорова-Смирнова
    kstest_result = kstest(wine_df[column], 'norm')

    # Вывод результата
    if kstest_result.pvalue > 0.05:
        print(f"{column} - Нормальное распределение")
    else:
        print(f"{column} - Ненормальное распределение")
```
